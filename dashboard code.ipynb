```python
schools, wins = np.unique(win_seed_sort['win_school_ncaa'], return_counts = True)
```

```python
shl_wins = pd.DataFrame(schools, wins).reset_index()
shl_wins.columns = ['number_wins', 'school']
shl_wins_gt50 = shl_wins[shl_wins['number_wins'] > 50]
```
